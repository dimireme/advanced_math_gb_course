```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5, 5, 21)
y = 3 * x + 1
y2 = (-1/3) * x + 1
plt.plot(x, y)
plt.plot(x, y2)
plt.grid()
plt.draw()
```

Прямые не кажутся перпендикулярными, так как масштабы осей не совпадают. Исправим 

```python
plt.plot(x, y)
plt.plot(x, y2)
plt.xlim((-5, 5))
plt.ylim((-4, 9))
plt.gca().set_aspect('equal', adjustable='box')
plt.grid()
plt.draw()
```

## Графики на плоскости

1. Нарисуйте график функции $y(x) = k\cos(x – a) + b$ для некоторых (2-3 различных) значений параметров k, a, b.


```python
x = np.linspace(-2*np.pi, 2*np.pi, 201)

y0 = np.cos(x)
y1 = np.cos(x - np.pi / 3) + 1
y2 = 0.5 * np.cos(x) - 1

plt.plot(x, y0, x, y1, x, y2)
plt.legend(['cos(x)', 'cos(x - pi/3) + 1', '0.5 * cos(x) - 1'])
plt.plot([np.pi / 3, np.pi / 3], [-1.5, 2], color="blue", linewidth=1, linestyle="--")
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.draw()
```
