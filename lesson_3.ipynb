2\.Почему прямые не кажутся перпендикулярными? (см.ролик)

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5, 5, 21)
y = 3 * x + 1
y2 = (-1/3) * x + 1
plt.plot(x, y)
plt.plot(x, y2)
plt.grid()
```

Прямые не кажутся перпендикулярными, так как масштабы осей не совпадают. Исправим это.

```python
plt.plot(x, y)
plt.plot(x, y2)
plt.xlim((-5, 5))
plt.ylim((-4, 9))
plt.gca().set_aspect('equal', adjustable='box')
plt.grid()
```

3\. Напишите код на Python, реализующий построение графиков окружности, эллипса, гиперболы.

```python
# график окружности
R = 10
x = np.linspace(-R, R, 1000)
y1 = np.sqrt(R**2 - x**2)
y2 = -np.sqrt(R**2 - x**2)
plt.axis('equal')
plt.plot(x, y1, 'royalblue', x, y2, 'royalblue')
plt.show()
```

```python
# график эллипса
a = 10
b = 6
x = np.linspace(-a, a, 1000)
y1 = b*np.sqrt(1 - (x/a)**2)
y2 = -b*np.sqrt(1 - (x/a)**2)
plt.axis('equal')
plt.plot(x, y1, 'royalblue', x, y2, 'royalblue')
plt.show()
```

```python
# график гиперболы
a = 3
b = 2
x = np.linspace(-20, 20, 1000)
X, Y = np.meshgrid(x, x)
plt.contour(X, Y, (X/a)**2-(Y/b)**2-1, [0])
plt.axis('equal')
plt.show()
```

3.1. Напишите код, который будет переводить полярные координаты в декартовы.

```python
def polar_to_dekart(R, s):
    x = R * np.cos(s)
    y = R * np.sin(s)
    return (x, y)
```

```python
polar_to_dekart(10, np.pi/3)
```

```python
polar_to_dekart(10, np.pi*2/3)
```

```python
polar_to_dekart(10, np.pi*5/4)
```
