3\. Напишите код на Python, реализующий построение графиков окружности, эллипса, гиперболы.

```python
# график окружности
R = 10
x = np.linspace(-R, R, 1000)
y1 = np.sqrt(R**2 - x**2)
y2 = -np.sqrt(R**2 - x**2)
plt.figure(figsize=(5, 5))
plt.axis('equal')
plt.grid()
plt.plot(x, y1, 'royalblue', x, y2, 'royalblue')
plt.show()
```

```python
# график эллипса
a = 10
b = 6
x = np.linspace(-a, a, 1000)
y1 = b*np.sqrt(1 - (x/a)**2)
y2 = -b*np.sqrt(1 - (x/a)**2)
plt.figure(figsize=(5, 5))
plt.axis('equal')
plt.grid()
plt.plot(x, y1, 'royalblue', x, y2, 'royalblue')
plt.show()
```

```python
# график гиперболы
a = 3
b = 2
x = np.linspace(-20, 20, 1000)
X, Y = np.meshgrid(x, x)
plt.figure(figsize=(5, 5))
plt.axis('equal')
plt.grid()
plt.contour(X, Y, (X/a)**2-(Y/b)**2-1, [0])
plt.show()
```

4.2. Пусть задана плоскость $A_1x + B_1y + C_1z + D_1 = 0$ и прямая ${{x-x_1}\over{x_2-x_1}} = {{y-y_1}\over{y_2-y_1}} = {{z-z_1}\over{z_2-z_1}}$. Как узнать, принадлежит прямая плоскости или нет?

В уравнении прямой точки $(x_1, y_1, z_1)$ и $(x_2, y_2, z_2)$ лежат на этой прямой. Если они обе принадлежат плоскости (являются решениями для уравнения плоскости), значит и вся прямая лежит на этой плоскости:

$$
\begin{equation*}
 \begin{cases}
   A_1x_1 + B_1y_1 + C_1z_1 + D_1 = 0
   \\
   A_1x_2 + B_1y_2 + C_1z_2 + D_1 = 0
 \end{cases}
\end{equation*}
$$

Если оба равенства верны, то прямая лежит на плоскости.

2\. Докажите, что при ортогональном преобразовании сохраняется расстояние между точками.

Пусть исходные точки заданы координатами $(x_1,y_1) (x_2, y_2)$. Квадрат расстояния между ними:

$$S^2 = (x_2-x_1)^2 + (y_2-y_1)^2$$ 

Обозначим $\Delta{x} = x_2-x_1, \Delta{y} = y_2-y_1$. Применим ортогональное преобразование:

$$
x_1^\prime = a_{11}x_1 + a_{12}y_1 + a_{13}\\
x_2^\prime = a_{11}x_2 + a_{12}y_2 + a_{13}\\
y_1^\prime = a_{21}x_1 + a_{22}y_1 + a_{23}\\
y_2^\prime = a_{21}x_2 + a_{22}y_2 + a_{23}
$$

Квадрат нового расстояния между точками:

$$
S^{\prime2} = (x_2^\prime - x_1^\prime)^2 + (y_2^\prime - y_1^\prime)^2 =\\
(a_{11}x_2 + a_{12}y_2 + a_{13} - a_{11}x_1 - a_{12}y_1 - a_{13})^2 +\\
(a_{21}x_2 + a_{22}y_2 + a_{23} - a_{21}x_1 - a_{22}y_1 - a_{23})^2 =\\
(a_{11}\Delta{x} + a_{12}\Delta{y})^2 + (a_{21}\Delta{x} + a_{22}\Delta{y})^2 =\\
a_{11}^2\Delta{x}^2 + 2a_{11}\Delta{x}a_{12}\Delta{y} + a_{12}^2\Delta{y}^2 +\\
a_{21}^2\Delta{x}^2 + 2a_{21}\Delta{x}a_{22}\Delta{y} + a_{22}^2\Delta{y}^2 = \\
(a_{11}^2 + a_{21}^2)\Delta{x}^2 + 2(a_{11}a_{12} + a_{21}a_{22})\Delta{x}\Delta{y} + (a_{12}^2 + a_{22}^2)\Delta{y}^2
$$

Используем свойства ортогонального преобразования:

$$
a_{11}^2 + a_{21}^2 = 1\\
a_{12}^2 + a_{22}^2 = 1\\
a_{11}a_{12} + a_{21}a_{22} = 0
$$

Тогда $S^{\prime2} = \Delta{x}^2 + \Delta{y}^2 = S^2$. То есть квадрат расстояния между двумя точками не изменился. Значит и расстояние между точками осталось преждним.

3.1. Напишите код, который будет переводить полярные координаты в декартовы.

```python
def polar_to_dekart(R, s):
    x = R * np.cos(s)
    y = R * np.sin(s)
    return (x, y)
```

```python
polar_to_dekart(10, np.pi/3)
```

```python
polar_to_dekart(10, np.pi*2/3)
```

```python
polar_to_dekart(10, np.pi*5/4)
```
