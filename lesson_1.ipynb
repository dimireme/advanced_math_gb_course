4. Постройте на одном графике две кривые y(x) для функции двух переменной y(k,x)=cos(k∙x), взяв для одной кривой значение k=1, а для другой – любое другое k, не равное 1.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5, 5, 201)
plt.plot(x, np.cos(x), label="cos(x)")
plt.plot(x, np.cos(3*x), label="cos(3x)")
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```
