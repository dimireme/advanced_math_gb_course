## Элементы теории вероятностей

1\. Найти вероятность выпадения 2 или 5 очков при подбрасывании игральной кости, на гранях которой имеются соответственно 1,2,3,4,5 и 6 очков.

$P(2) = P(5) = 1/6$. Так как события несовместные, то $P(2+5) = P(2) + P(5) = 1/3$.

2\. Найти вероятность того, что при двух подбрасываниях той же самой игральной кости сначала выпадет 2, а затем 5.

$P = P(2) \cdot P(5) = 1/36$

3\. Найти вероятность выпадения 2 и 5 очков при двух подбрасываниях той же самой игральной игральной кости.

В этом случае из 36 исходов положительными будут два: (2, 5) и (5, 2). $P = 1/18$.

4\. На отрезке AB длиной 20 см наугад отметили точку C. Какова вероятность, что она находится на расстоянии не более 9 см от точки А и не более 15 см от точки В?

Успешный исход будет, если точка С попадёт на отрезок длиной 4 см. $P = 4/20 = 0.2$.

5\. Телефонный номер состоит из 7 цифр. Какова вероятность, что это номер 8882227?

Если считать что номер не может начинаться с 0, то всего номеров может быть $9 \cdot 10^6$. Искомая вероятность $P = 1 / (9 \cdot 10^6)$ 

6\. Набирая номер телефона, абонент забыл 2 последние цифры, и, помня только то, что эти цифры различны и среди них нет нуля, стал набирать их наудачу. Сколько вариантов ему надо перебрать, чтобы наверняка найти нужный номер? Какова вероятность того, что он угадает номер с первого раза?

Потребуется $9 \cdot 8 = 72$ попыток. Шанс угадать с первой попытки $P = 1/72$.

7\. Чёрный куб покрасили снаружи белой краской, затем разрезали на 27 одинаковых маленьких кубиков и как попало сложили из них большой куб. С какой вероятностью все грани этого куба будут белыми?

Один маленький кубик можно расположить в пространстве $6 \cdot 4 = 24$ способами (одна грань из шести нижняя, плюс поворот относительно вертикальной оси). Всего собрать большой куб можно $N = 27! \cdot 24^{27}$ способами. Начнём собирать наш большой куб со всеми наружними белыми гранями. 

Центральный кубик. В качестве центрального кубика мы можем выбрать только один из 27 имеющихся. И расположить его можно 24 способами. Все они будут подходящими.

Кубики в центре граней. Для первой грани мы можем выбрать один из 6 кубиков и расположить его 4 способами. Для второй - один из 5 оставшихся и так же расположить его 4 способами. И т.д. Всего вариантов выбрать шесть кубиков в центрах грани $6! \ cdot 4^6$

Кубики на границе граней. Их всего 12 и при расположении успешными будут две ориентации каждого кубика в пространстве. Аналогично предыдущему пункту, получим $12! \cdot 2^{12}$ способов успешно выставить кубики на границе граней.

Кубики по углам куба. Их 8, а способов ориентации 3. Всего получаем $8! \cdot 3^8$ способов успешно расположить кубики по углам куба. 


В итоге получим $P = \frac{24 \cdot (6! \cdot 4^6)(12! \cdot 2^{12})(8! \cdot 3^8)}{27! \cdot 24^{27}}$
